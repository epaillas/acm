# `prepare_sim` control plots

This notebook contains the control plots to check the completeness of the Dark Matter catalogs produced by `abacusnbody.hod.prepare_sim`.

```python
from pathlib import Path
from acm.utils.default import cosmo_list as cosmo_targets
from acm.utils.abacus import get_abacus_phases
from utils import find_mocks, get_file_count
```

```python
abacus_dir = Path('/global/cfs/cdirs/desi/cosmosim/Abacus/small/')
box_dir = Path('/pscratch/sd/s/sbouchar/summit_subsamples/boxes')
z = 0.2
phase_targets = get_abacus_phases(abacus_dir, z=0.2)[1]
```

## 1- Base box catalogs

### 1.1 - Available cosmologies

```python
phase = 0
files = find_mocks(box_dir, f'base/AbacusSummit_base_c*_ph{phase:03d}')

available = [int(f.split('_')[-2].lstrip('c')) for f in files]
missing = set(cosmo_targets) - set(available)
extras = set(available) - set(cosmo_targets)

print(f'Found {len(available)} available cosmologies')
if missing:
    print(f'Missing cosmologies: {sorted(list(missing))}')
else:
    print('All cosmologies are available!')
if extras:
    print(f'Extra cosmologies: {sorted(list(extras))}')
```

### 1.2 - Mock completeness check

Checking that the number of halo and particle files per cosmology is as expected.

```python
halo_counts, particle_counts = get_file_count(files, z, indexes=available)

expected_file_number = 34

for cosmo in cosmo_targets:
    hc = halo_counts.get(cosmo, 0)
    pc = particle_counts.get(cosmo, 0)
    if hc != expected_file_number or pc != expected_file_number:
        print(f'Cosmology {cosmo:03d} has {hc} halo files and {pc} particle files (expected {expected_file_number})')
```

## 2- Small boxes catalogs

### 2.1 - Available phases

```python
cosmo = 0
files = find_mocks(box_dir, f'small/AbacusSummit_small_c{cosmo:03d}_ph*')

available = [int(f.split('_')[-1].lstrip('ph')) for f in files]
missing = set(phase_targets) - set(available)
extras = set(available) - set(phase_targets)

print(f'Found {len(available)} available phases')
if missing:
    print(f'Missing phases: {sorted(list(missing))}')
else:
    print('All phases are available!')
if extras:
    print(f'Extra phases: {sorted(list(extras))}')
```

### 2.2 - Mock completeness check

Checking that the number of halo and particle files per phase is as expected.

```python
halo_counts, particle_counts = get_file_count(files, z, indexes=available)

expected_file_number = 1

for phase in phase_targets:
    hc = halo_counts.get(phase, 0)
    pc = particle_counts.get(phase, 0)
    if hc != expected_file_number or pc != expected_file_number:
        print(f'Phase {phase:03d} has {hc} halo files and {pc} particle files (expected {expected_file_number})')
```
