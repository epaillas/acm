# AbacusSummit small boxes phase numbers

This notebook is used to figure out the available phase numbers for the small boxes of the AbacusSummit simulation suite and convert this list to a bash array for later use in job submission scripts.

For more information, see the [covariance small boxes](https://abacussummit.readthedocs.io/en/latest/simulations.html#covariance-small-boxes) information page.

```python
from pathlib import Path
from acm.utils.abacus import get_abacus_phases
from utils import list_to_sequence
```

```python
z = 0.2
abacus_dir = Path('/global/cfs/cdirs/desi/cosmosim/Abacus/small/')

abacus_fns, phases = get_abacus_phases(abacus_dir, z)
print(f'Found {len(abacus_fns)} files at z={z:.3f}')
```

```python
phase_sequence = list_to_sequence(phases)

stp = [] # sequence to print
for item in phase_sequence:
    if isinstance(item, tuple):
        stp.append(f'{{{item[0]}..{item[1]}}}')
    else:
        stp.append(str(item))

stp = ' '.join(stp)
print('BASH array format:')
print(f'PHASE_LIST=({stp})')
```
