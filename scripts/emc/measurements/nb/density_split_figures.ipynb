```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'


data_dir = '/pscratch/sd/e/epaillas/emc/v1.2/abacus/raw_measurements/density_split/c002_ph000/seed0/'
data_fn = Path(data_dir) / 'dsc_xiqg_poles_c002_hod077.npy'
data = np.load(data_fn, allow_pickle=True)

fig, ax = plt.subplots(1, 2, figsize=(8, 3))

for quantile in [0, 1, 3, 4]:
    s, poles = data[quantile][::4](ells=(0, 2), return_sep=True)

    for ell in [0, 2]:
        ax[ell // 2].plot(s, s ** 2 * poles[ell // 2], label=f'quantile {quantile}')

ax[0].set_ylabel(r'$s^2 \xi_0(s)\, [h^{-2}{\rm Mpc}^2]$')
ax[1].set_ylabel(r'$s^2 \xi_2(s)\, [h^{-2}{\rm Mpc}^2]$')
ax[0].set_xlabel(r'$s\, [h^{-1}{\rm Mpc}]$')
ax[1].set_xlabel(r'$s\, [h^{-1}{\rm Mpc}]$')
plt.tight_layout()
plt.savefig('dsc_xiqg_poles_c000_hod077.png', dpi=300, bbox_inches='tight')
plt.show()
```
