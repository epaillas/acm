```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'


data_dir = '/pscratch/sd/e/epaillas/emc/v1.2/abacus/raw_measurements/density_split/c000_ph000/seed0/'
data_fn = Path(data_dir) / 'wst_c000_hod199.npy'
data = np.load(data_fn, allow_pickle=True)

fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(data)

ax.set_xlabel('bin index')
ax.set_ylabel('WST coefficient')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

mesh_acm = np.load('/pscratch/sd/e/epaillas/emc/debug/abacus/raw_measurements/base_estimator/c000_ph000/seed0/delta_mesh_c000_hod008.npy')
mesh_acm[mesh_acm > 5] = np.nan

mesh_georgios = np.load('/pscratch/sd/g/gvalogia/density_200_CIC.npy').reshape(-1)
mesh_georgios[mesh_georgios > 5] = np.nan

fig, ax = plt.subplots(figsize=(4, 3))

ax.hist(mesh_acm, bins=50, alpha=0.5, label='ACM', density=True)
ax.hist(mesh_georgios, bins=50, alpha=0.5, label='Georgios', density=True)

ax.legend()
ax.set_xlabel(r'$\delta$')
ax.set_ylabel('PDF')
ax.set_xlim(-1, 6)
plt.tight_layout()
plt.savefig('mesh_comparison.pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'


wst_enrique = np.load(
    '/pscratch/sd/e/epaillas/emc/wst_from_acm_mesh.npy',
    allow_pickle=True
)
wst_georgios = np.load('/pscratch/sd/e/epaillas/emc/wst_from_georgios_mesh.npy',
    allow_pickle=True
)

fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(wst_enrique, label='Enrique')
ax.plot(wst_georgios, label='Georgios', linestyle='dashed')

ax.set_xlabel('bin index')
ax.set_ylabel('WST coefficient')
ax.legend()
plt.tight_layout()
plt.show()
```
