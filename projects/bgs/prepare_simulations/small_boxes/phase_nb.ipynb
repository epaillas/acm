```python
sim_path = Path('/global/cfs/cdirs/desi/cosmosim/Abacus/small/')

simu_list = list(sorted(sim_path.glob('AbacusSummit_small_c000_ph*')))

sim_at_z0200 = []
for sim in simu_list:
    path = sim / 'halos'
    search = list(path.glob('*z0.200'))
    if len(search) > 0:
        sim_at_z0200.append(str(sim)[-4:])

np.save('AbacusSmall_phases.npy', sim_at_z0200)
print(len(sim_at_z0200), 'small boxes for z=0.2')
```

```python
# Identify sequences within the list

sim_at_z0200 = sorted(sim_at_z0200)

seq = []
for i, ph in enumerate(sim_at_z0200):
    if len(seq) == 0:
        seq.append(ph)
    if i == len(sim_at_z0200)-1:
        seq.append(ph)
        break
    
    if sim_at_z0200[i+1] == str(int(ph)+1):
        # print(sim_at_z0200[i+1], str(int(ph)+1))
        continue
    else:
        seq.append(ph)
        seq.append(sim_at_z0200[i+1])
    
# Pair sequences

sequences = [[seq[i], seq[i+1]] for i in range(0, len(seq)-1, 2)]

# Format it to a bash list
for s in sequences:
    if s[0]==s[1]:
        sequences[sequences.index(s)] = int(s[0])
    else:
        sequences[sequences.index(s)] = f'{{{s[0]}..{s[1]}}}' 

sequences = ' '.join(map(str, sequences))

print('Sequences:\n', sequences)
```
