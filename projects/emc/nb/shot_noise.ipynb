```python
from pathlib import Path
import matplotlib.pyplot as plt
from pycorr import TwoPointCorrelationFunction
import numpy as np
%matplotlib inline

hods = [0, 1, 2, 3, 4, 5]

fig, ax = plt.subplots()
seeds = [1]

for hod in hods:
    for seed in seeds:
        data_dir = f'/pscratch/sd/e/epaillas/emc/training_sets/test/tpcf/z0.5/yuan23_prior/c000_ph000/seed{seed}/'
        data_fn = Path(data_dir) / f'tpcf_hod{hod:03}.npy'
        data = TwoPointCorrelationFunction.load(data_fn)[::4]
        # data.select((0, 30))

        s, multipoles = data(ells=(0, 2), return_sep=True)
        multipoles = np.concatenate(multipoles)

        # ax.plot(s, s**0*multipoles[0])
        ax.plot(multipoles)
    
# ax.set_yscale('log')
plt.show()
```
