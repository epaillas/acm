```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

fig, ax = plt.subplots(figsize=(10, 4))

y = []
for phase in range(3000, 5000):

    data_dir = '/global/homes/e/epaillas/pscratch/emc/covariance_sets/knn/z0.5/yuan23_prior/'
    data_fn = Path(data_dir) / f'knn_ph{phase:04}_hod466.npy'
    if not data_fn.exists(): continue
    data = np.load(data_fn).reshape(-1)
    y.append(data)
    ax.plot(data)
plt.show()

y = np.array(y)

cov = np.cov(y, rowvar=False) / 64

fig, ax = plt.subplots()

ax.plot(np.sqrt(np.diag(cov)))

plt.show()

np.linalg.inv(cov)
```
